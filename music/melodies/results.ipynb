```python
from music.lib import *
from music.melodies.eval import *
from music.melodies.query import get_output_file
```

```python
gpt4 = 'gpt-4-0314'
gpt3 = 'gpt-3.5-turbo-0301'
claude = 'claude-v1.3'
palm2 = 'text-bison-001'
```

```python
MODELS = [gpt4, gpt3, claude, palm2,]
```

```python
EXP_DIR = 'music/melodies/output'
```

```python
all_results = load_all_results(EXP_DIR, MODELS)
```

```python
def print_results(all_results, model):
    results = []
    num_examples = []
    non_cmajor_keys = [k for k in all_results['key'].unique() if k != 'C major']
    keys = [['C major'], non_cmajor_keys]

    for control in [False, True]:
        for cot in [False, True]:
            
            for sub_keys in keys:
    
                control_df = all_results[all_results['is_control']==control]
                model_df = control_df[control_df['model']==model]

                cot_df = model_df[model_df['chain_of_thought']==cot]

                subset = cot_df[cot_df['is_control']==control]
                subset = subset[subset['key'].isin(sub_keys)]
                acc, num_correct, num_total = get_subset_acc(subset,)

                results.append(acc)
                num_examples.append(num_total)

    str_results = " & ".join([str(round(r, 1)) for r in results])
    num_examples = " & ".join([str(n) for n in num_examples])

    return str_results, num_examples
```

```python
# Results for melody retrieval task
for idx, model in enumerate(MODELS):
    str_results, num_examples = print_results(all_results, model)
    
    if idx == 0:
        print(num_examples)
        print()
    print(str_results)
```

```python
# GPT4: results by n
for idx, n in enumerate(range(7)):
    sub_df = all_results[all_results['note']==n]
    str_results, num_examples = print_results(sub_df, gpt4)
    if idx == 0:
        print(num_examples)
        print()
    print(f'{n+1} & ' + str_results + '\\\\')
```
