```python
def print_results(all_results, model, tunings, control, chain_of_thought=None):
    
    if chain_of_thought is None:
        chain_of_thought = [False, True]
    
    control_df = all_results[all_results['is_control']==control]
    model_df = control_df[control_df['model']==model]
    results = []
    num_examples = []
    
    for cot in chain_of_thought:

        for tuning in tunings:
        
            cot_df = model_df[model_df['chain_of_thought']==cot]
        
            tuning_df = cot_df[cot_df['tunings']==tuning]
            if control:
                acc, num_correct, num_total = get_subset_control_acc(tuning_df, len(tuning))
            else:
                acc, num_correct, num_total = get_subset_acc(tuning_df,)

            results.append(acc)
            num_examples.append(num_total)
        
    str_results = " & ".join([str(round(r, 1)) for r in results])
    num_examples = " & ".join([str(n) for n in num_examples])
    
    return str_results, num_examples
```

```python
def print_all_results(results, models, tunings, is_control):
    print(tunings)

    for idx, model in enumerate(models):
        str_results, num_examples = print_results(results, model, 
                                                  tunings, 
                                                  is_control, chain_of_thought=None)
        if idx == 0:
            print(num_examples)
        print(str_results)
```

```python
# Guitar: task results
print_all_results(guitar_results, MODELS, guitar_tunings, False)
```

```python
# Guitar: CCC results
print_all_results(guitar_results, MODELS, guitar_tunings, True)
```

```python
# Guitar: task results by chords

is_control = False
print(guitar_tunings)
chord_types = ['major', 'minor', '5', '7',  '6', 'sus4', 'dim7', 'aug7', 'sus2', 'm7']
for idx, chord in enumerate(chord_types):

        
    chord_results = guitar_results[guitar_results['chord_type']==chord]
    
    str_results, num_examples = print_results(chord_results, gpt4, guitar_tunings, 
                  is_control, chain_of_thought=None)

    if idx == 0:
        print(num_examples)    
    print(chord + " & " + str_results + " \\\\")
```
